```python
import random

num_valores = 10000
valores = [random.randint(0, 1) for _ in range(num_valores)]
frecuencia_absoluta = {0:valores.count(0), 1: valores.count(1)}
frecuencia_relativa = {0: frecuencia_absoluta[0] / num_valores, 1: frecuencia_absoluta[1] / num_valores}
print("valores generados", valores)
print("frecuencia absoluta de 0", frecuencia_absoluta[0])
print("frecuencia absoluta de 1", frecuencia_absoluta[1])
print("frecuencia relativa de 0", frecuencia_relativa[0])
print("frecuencia relativa de 0", frecuencia_relativa[1])
```

```python
import random
import matplotlib.pyplot as plt

num_valores = 1000
valores = [random.randint(0, 1) for _ in range(num_valores)]

constant_values = [0.5 if i % 2 == 0 else 0.7 for i in range(num_valores)]

plt.plot(valores, label='valores aleatorios', color='blue')
plt.plot(constant_values, label='valor constante', color='red')
plt.ylabel('indice')
plt.xlabel('valores')
plt.title('grafico de valores aleatorios con valor constante')
plt.legend()
plt.grid(True)
plt.show()
```
